```python
import pandas as pd
```

```python
data = pd.read_csv('data_set.csv')
```

```python
data.describe()
```

```python
data.columns
```

```python
data['Date'] = pd.to_datetime(data['date_time'], errors='coerce')
```

```python
data['year'] = data['Date'].dt.year
data['month'] = data['Date'].dt.month
data['hour'] = data['Date'].dt.hour
```

```python
data.head()
```

```python
data.replace(to_replace='None', value=0)
```

```python
def transform_holiday(holiday):
    if holiday == 'None':
        return 0
    else:
        return 1
```

```python
data['holiday'] = data['holiday'].apply(transform_holiday)
```

```python
data.loc[126]
```

```python
data.describe()
```

```python
y = data.traffic_volume
X = data.drop(['traffic_volume','year','weather_description','date_time','Date'], axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
# Divide data into training and validation subsets
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import OneHotEncoder
s = (X_train.dtypes == 'object')
object_cols = list(s[s].index)
# Apply one-hot encoder to each column with categorical data
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X_train[object_cols]))
OH_cols_valid = pd.DataFrame(OH_encoder.transform(X_valid[object_cols]))

# One-hot encoding removed index; put it back
OH_cols_train.index = X_train.index
OH_cols_valid.index = X_valid.index

# Remove categorical columns (will replace with one-hot encoding)
num_X_train = X_train.drop(object_cols, axis=1)
num_X_valid = X_valid.drop(object_cols, axis=1)

# Add one-hot encoded columns to numerical features
X_train = pd.concat([num_X_train, OH_cols_train], axis=1)
X_valid = pd.concat([num_X_valid, OH_cols_valid], axis=1)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
rf_model = RandomForestRegressor( random_state=1)
rf_model.fit(X_train,y_train)
rf_val_mae = mean_absolute_error(rf_model.predict(X_valid),y_valid)
print("Validation MAE for Random Forest Model: {}".format(rf_val_mae))
```
